```python
import matplotlib.pyplot as plt

# Create sample data
x = range(10)
y = [i**2 for i in x]

# Create a plot
fig, ax = plt.subplots()

# Plot the data
ax.plot(x, y, label='Quadratic')

# Add a title to the plot
ax.set_title('Sample Plot')

# Add labels to the x and y axes
ax.set_xlabel('X-axis Label')
ax.set_ylabel('Y-axis Label')

# Customize axis ticks
ax.set_xticks(range(0, 11, 2))
ax.set_yticks(range(0, 101, 20))
ax.set_xticklabels(['zero', 'two', 'four', 'six', 'eight', 'ten'])
ax.set_yticklabels(['0', '20', '40', '60', '80', '100'])

# Add a legend
ax.legend()

# Add grid lines
ax.grid(True)

# Add annotations
ax.annotate('Maximum', xy=(9, 81), xytext=(5, 90),
            arrowprops=dict(facecolor='black', shrink=0.05))

# Add a figure title for multi-plot figures (optional)
fig.suptitle('Figure Title', fontsize=16)

# Display the plot
plt.show()
```

```python
import numpy as np
x = np.linspace(0,3, 100)
y = [np.log(i) for i in x]

plt.plot(x,y, label="logarithmic")
plt.plot(x, np.exp(x), label = "Exponential")
plt.title("y vs x", fontsize=5)
plt.suptitle("Comparison of plots")
plt.legend()
```
