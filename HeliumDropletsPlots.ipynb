```python
import numpy as np, matplotlib.pyplot as plt


filearray = [0,1,2,3,4] # file indices of interest here

labbookdata = np.array(
    [
        [0, 6000, 3500, 0, -1750, 0.20, None, "Closed"],
        [1, 6000, 3500, 0, -1750, 0.20, None, "Open"], 
        [2, 6000, 3500, 0, -1750, 0.05, None, "Open"],
        [3, 6000, 3500, 0, -1750, 0.10, None, "Open"],
        [4, 6000, 3500, 0, -1750, 0.05, None, "Closed"],
    ])

def spectroParamsExtract(filepath):
    with open(filepath, 'r') as file:
        file_contents = file.read()
        
    file_contents = file_contents.split("\n") 
    file_contents = file_contents[0] 
    file_contents = file_contents.split() 

    spectrometerData = np.array([float(num) for num in file_contents]) 
    
    return spectrometerData # exporting the number array


from matplotlib.backends.backend_pdf import PdfPages
with PdfPages("Day2_He_droplets.pdf") as pdf:
    
    k = 0 
    
    for i in filearray: 
        spectroParamFile = f"./Lab Data/Data_240620/rdm24-06-20_{i}.txt"
        datafile = f'./Lab Data/Data_240620/24-06-20_{i}_ch2.txt'
        

        spectrometerParameters = spectroParamsExtract(spectroParamFile)
        
        data = np.loadtxt(datafile, delimiter=' ');
        timeDelay = data[:, 0]
        intensity = data[:, 1]
        
        # plotting the data
        plt.plot(timeDelay,intensity, label=f"He Droplets, power={labbookdata[k][5]*100}%, Valve = {labbookdata[k][7].upper()}", linewidth = 0.2, color="blue")
        plt.xlabel("Time Delay (in micro-s)", fontsize=8)
        plt.ylabel("Signal Intensity - Detector (mV) ", fontsize=8)
        plt.legend()
        
        vgradient = f"{(int)(labbookdata[k][1])} {(int)(labbookdata[k][2])} {(int)(labbookdata[k][3])} {(int)(labbookdata[k][4])}"
        k+=1
        
        plt.title(f"Laser Iterations {spectrometerParameters[0]}; Voltage Gradient : "+vgradient, fontsize=8)
        plt.suptitle(f"File #{i}; Helium Droplet spectrum at various powers, T = 9.5K ", fontsize=12)
        
        pdf.savefig()
        plt.close()
        
```
